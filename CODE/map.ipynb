1. obtain data

```python
import pickle
def read_pkl(path):
    df=open(path,'rb')
    data3=pickle.load(df)
    return data3

map_data = pd.DataFrame(columns =['state', 'polarity', 'date'])
idx = 0
for m in range(6, 10):
    max_day = 31
    if m==6 or m==9: max_day=30
    for d in range(1, max_day+1):
        # transactions['date_formatted']=pd.to_datetime(transactions['date'], format='%d.%m.%Y')
        date = "20210" + str(m)
        day = "2021_0"+str(m)+"_"
        if d < 10:
            day = day + "0" + str(d)
            date = date + "0" + str(d)
        else:
            day += str(d)
            date += str(d)
        df = pd.read_pickle("data/"+day + ".pkl")

        state_polarity = {}
        for i in range(df.shape[0]):
            state_name = df.iloc[i]["state"]
            blob = TextBlob(str(df[i:i+1]["text"]))
            state_polarity[state_name] = blob.sentiment[0]
        for key in state_polarity:
            map_data.loc[idx]={'state':key, 'polarity': state_polarity[key], 'date':date}
            idx += 1
        print(day)
```
